```python
%pip install pandas numpy
import pandas as pd
import numpy as np


# Split into 4 equal parts
split_articles = np.array_split(articles, 4)

# Access individual pieces
piece_1 = split_articles[0]
piece_2 = split_articles[1]
piece_3 = split_articles[2]
piece_4 = split_articles[3]

# Check the sizes of each part
for i, piece in enumerate(split_articles, 1):
    print(f"Piece {i}: {len(piece)} rows")
```

## Test stance detection

```python
# Extract the first two rows
two_row_piece = articles.iloc[:10]

# Display the new piece
two_row_piece
```

```python
# Initialize the zero-shot classification pipeline
zeroshot_classifier = pipeline("zero-shot-classification", model="MoritzLaurer/deberta-v3-large-zeroshot-v1.1-all-33")

# Set up the classification parameters
hypothesis_template = "This article is {} abortion"
classes_verbalized = ["against", "for", "neutral toward"]

# Assuming CNN_Fox_joined_17_05 is your DataFrame and "body" contains the articles and "textID" contains the IDs
def classify_article(text):
    # Run the zero-shot classifier on each article
    output = zeroshot_classifier(text, classes_verbalized, hypothesis_template=hypothesis_template, multi_label=False)
    # Create a dictionary with class labels as keys and their confidence scores as values
    return {label: score for label, score in zip(output['labels'], output['scores'])}

# Apply the classification to each row in the "body" column and collect the results
stance_detection_results = piece_4['body'].apply(classify_article)

# Convert the results into a new DataFrame, with each article as a row and classes as columns
stance_detection_piece_4 = pd.DataFrame(stance_detection_results.tolist())

# Add the textID column, source column the original article text from the original DataFrame
stance_detection_piece_4.insert(0, 'textID', piece_4['textID'].values)
stance_detection_piece_4.insert(1, 'date', piece_4['date'].values)
stance_detection_piece_4.insert(2, 'source', piece_4['source'].values)
stance_detection_piece_4.insert(3, 'body', piece_4['body'].values)
stance_detection_piece_4.insert(4, 'leaning', piece_4['leaning'].values)


# Add a new column 'stance' which contains the class with the highest probability
stance_detection_piece_4['stance'] = stance_detection_piece_4[['for', 'neutral toward', 'against']].idxmax(axis=1)

# Display the new DataFrame to the user
stance_detection_piece_4
```

```python
# save as csv

# Define the file path where you want to save the CSV🔥
output_file_path = "/work/Bachelor/stance_detection_piece_4.csv"

# Save the filtered DataFrame as a CSV file
stance_detection_piece_4.to_csv(output_file_path, index=False, encoding='utf-8')
```
